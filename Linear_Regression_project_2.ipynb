```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
df = pd.read_csv('Power Plant Data.csv')
```

```python
df.head()
```

```python
df = df.drop(['V','AP','RH'], axis=1)
```

```python
df.head()
```

```python
x = df.iloc[:,0]
y = df.iloc[:,1]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# building the model
m = 0
c = 0

L = 0.001
epochs = 1000

n = float(len(x))   # gragdiant desent proses
for i in range(len(x)):
  y_pred = m*x + c
  d_m = (-2/n) * sum(x * (y - y_pred))  # formula
  d_c = (-2/n) * sum(y - y_pred)
  m = m - L * d_m
  c = c -L *d_c

print(m,c)

```

```python
# making prediction
y_pred = m*x + c
```

```python
plt.scatter(x,y)
plt.scatter(x,y_pred)
plt.show()
```

```python
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```

```python
# calculate the accuracy of the model
from sklearn.metrics import r2_score

r2 = r2_score(y, y_pred)
print(f'R-squared: {r2}','in',r2*100,'%')
```
