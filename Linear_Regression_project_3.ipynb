```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x='year',y='per capita income (US$)',data=df)
```

```python
x = df['year']
y = df['per capita income (US$)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(np.array(x_train).reshape(-1, 1), y_train)
```

```python
reg.predict([[2020],[ 2025]])
```

```python
plt.scatter(x, y, label='Original Data')
plt.plot(x, reg.predict(np.array(x).reshape(-1, 1)), color='red', label='Best Fit Line')
plt.scatter([2020, 2025], reg.predict([[2020], [2025]]), color='green', marker='x', s=100, label='Predicted Points')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income vs. Year')
plt.legend()
plt.show()
```
